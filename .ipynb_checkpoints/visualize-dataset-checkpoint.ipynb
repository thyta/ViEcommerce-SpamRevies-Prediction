# VISULIZATION

Percent of spam

```python
# count spam and no-spam
train_c_spam = data_train[data_train["label"] == 0].shape[0]
train_c_nospam = data_train[data_train["label"] == 1].shape[0]
train_data = [train_c_spam, train_c_nospam]

dev_c_spam = data_dev[data_dev["label"] == 0].shape[0]
dev_c_nospam = data_dev[data_train["label"] == 1].shape[0]
dev_data = [dev_c_spam, train_c_nospam]

test_spam = data_test[data_test["label"] == 0].shape[0]
test_nospam = data_test[data_test["label"] == 1].shape[0]
test_data = [test_spam, train_c_nospam]
```

```python
def visuallize_percent_spam (data):
    def percent_cal (pct, all_values):
        absolute = int(pct / 100.*np.sum(all_values))
        #f: %, d: sluong
        return "{:.1f}%\n({:d})".format(pct, absolute)

    # Creating plot
    fig, ax = plt.subplots(figsize =(10, 7))
    wedges, texts, autotexts = ax.pie (
                                    data,
                                    autopct = lambda pct: percent_cal(pct, data),
                                    explode = [.1, 0], # space between
                                    labels = ['Non-spam','Spam'],
                                    shadow = True,
                                    colors = ['green','yellow'],
                                    startangle = 90,
                                    wedgeprops = {'linewidth':1, 'edgecolor':'black'}, # border
                                    textprops = dict(color ="black"))

    # Adding legend
    ax.legend(wedges,  ['Non-spam','Spam'],
            title ="Label",
            loc ="center left",
            bbox_to_anchor =(1, 0, 0.5, 1))

    plt.setp(autotexts, size = 15, weight ="bold")
    ax.set_title("The percentage between Non-spam & Spam",fontsize=20, loc='center')

    # show plot
    plt.show()
```

```python
print("Visualize train data")
visuallize_percent_spam(train_data)
```

```python
print("Visualize dev data")
visuallize_percent_spam(dev_data)
```

```python
print("Visualize test data")
visuallize_percent_spam(test_data)
```

```python
data_train["comment"] = data_train["comment"].apply(ViTokenizer.tokenize)
data_train.head()
```

```python
# filder stop word
def filter_stop_words(train_sentences, stop_words):
    new_sent = [word for word in train_sentences.split() if word not in stop_words]
    train_sentences = ' '.join(new_sent)
    return train_sentences

# remove emoji
def deEmojify(text):
    regrex_pattern = re.compile(pattern = "["
        u"\U0001F600-\U0001F64F"  # emoticons
        u"\U0001F300-\U0001F5FF"  # symbols & pictographs
        u"\U0001F680-\U0001F6FF"  # transport & map symbols
        u"\U0001F1E0-\U0001F1FF"  # flags (iOS)
                           "]+", flags = re.UNICODE)
    return regrex_pattern.sub(r'',text)
# declare func and adding some remove
def preprocess(text, tokenized = True, lowercased = True):
    text = ViTokenizer.tokenize(text) if tokenized else text
    text = filter_stop_words(text, stopwords)
    text = deEmojify(text)
    text = text.lower() if lowercased else text

    text = text.strip()
    text = re.compile('<.*?>').sub('', text)
    text = re.sub('\s+', ' ', text) 
    text = re.sub(r'\[[0-9]*\]',' ',text)
    text = re.sub(r'[^\w\s]', '', text)
    text = re.sub(r'\d',' ',text)
    text = re.sub(r'\s+',' ',text)
    return text
```

```python
text_01 = "[HN] Hiện tại bên mình vẫn đang cần tuyển vị trí C# developer (.NET framework, C# biết winforms (2-3 năm)) sẵn sàng làm thêm FE hoặc Python, Java nếu được giao. Lương thỏa thuận, có 13-15 tháng lương. Quan tâm ib lấy JD."
text_02 = preprocess(text_01)
print("After preprocess: ", text_01, "\n")
print("Before preprocess: ", text_02)
```
